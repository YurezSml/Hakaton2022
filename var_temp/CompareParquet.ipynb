```python
import pandas as pd
```

```python
file_1 = "test_private_sample_submission_mfti.parquet"
file_2 = "test_private_sample_submission_mfti_old.parquet"
```

```python
def convert_pred_to_int(pred_list):
    return [int(x) for x in pred_list]
```

```python
df_1 = pd.read_parquet(file_1)
df_1['predictions'] = df_1['predictions'].apply(convert_pred_to_int)
df_1
```

```python
df_2 = pd.read_parquet(file_2)
df_2['predictions'] = df_2['predictions'].apply(convert_pred_to_int)
df_2
```

```python
df_common = pd.merge(df_1, df_2, on='cookie_id', sort=False)
df_common['predictions'] = df_common['predictions_x'] + df_common['predictions_y']
df_common = df_common.drop(['predictions_x','predictions_y'],axis=1)
df_common
```

```python
def get_num_diff_ids(ids_list):
    return len(ids_list) - len(set(ids_list))-5
```

```python
df_common['num_diff_ids'] = df_common['predictions'].apply(get_num_diff_ids)
df_common
```
